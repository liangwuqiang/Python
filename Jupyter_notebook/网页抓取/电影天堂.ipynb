```python
from urllib import request
from bs4 import BeautifulSoup
```

```python
url_dytt8 = 'http://www.dytt8.net/'
response = request.urlopen(url_dytt8)
html = response.read()
print(html)
```

```python
html = html.decode('gb2312')
print(html)
```

```python
soup = BeautifulSoup(html, 'html.parser')
# catalog = soup.select('.co_content8')
# catalog = soup.find_all(class_='co_content8')
# catalog = soup.select('.co_area2 .co_content8 .inddline')
co_area2 = soup.select('.co_area2')[1]
co_content8 = co_area2.select('.co_content8')[0]
print(co_content8)
```

```python
inddline = co_content8.select('.inddline a')
print(inddline)
```

```python
item = {}
for each in inddline:
    if each.text != '最新电影下载':
        item['title'] = each.text
        item['link'] = url_dytt8 + each['href']
        print(item['title'], item['link'])
```
