```python
import requests
from bs4 import BeautifulSoup
import hashlib
import re
import os
```

```python
url = r'http://www.codingpy.com/article/getting-started-with-jupyter-notebook-part-2/'
response = requests.get(url)
# response
html = response.text
print(html)
```

```python
soup = BeautifulSoup(html, 'html.parser')
title = soup.select('.header h1')[0].text.strip()
print(title)
```

```python
content = soup.select('.article-content')[0]
print(content)
```

```python
def md5(string):
    """ 将字符串转成哈希码 """
    import hashlib
    if not isinstance(string, str):
        string = str(string)
    md5 = hashlib.md5()  
    md5.update(string.encode('utf-8'))
    return md5.hexdigest()
```

```python
selectedImages = content.select('img')
# imagesSelected
for selectedImage in selectedImages:
    imageUrl = selectedImage['src']
#     print(imageUrl)
#     print(md5(imageUrl))
    imageMd5 = md5(imageUrl)
    print(imageUrl)
    if not os.path.exists('images'):
        os.mkdir('images')
    filename = './images/' + imageMd5 + os.path.splitext(imageUrl)[-1]
    print(filename)
#     imageResponse = requests.get(imageUrl)
#     print(imageResponse.status_code)
    if imageResponse.status_code == 200:
        print('aa')
        with open(filename, 'wb') as f:
            f.write(imageResponse.content)
            print('ok')
```
