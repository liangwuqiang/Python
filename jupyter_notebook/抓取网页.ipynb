```python
from urllib import request
url = 'http://www.codingpy.com/article/getting-started-with-jupyter-notebook-part-1/'
req = request.Request(url)
res = request.urlopen(req)
html = res.read().decode('utf-8')
print(html)
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(html, 'html.parser')
```

```python
title = soup.find_all(class_='header')
print(title)
```

```python
title = soup.select('.header h1')[0].text.strip()
print(title)
```

```python
content = soup.select('.article-content')[0]
print(content)
```

```python
def md5(string):
    """ 将字符串转成哈希码 """
    import hashlib
    if not isinstance(string, str):
        string = str(string)
    md5 = hashlib.md5()  
    md5.update(string.encode('utf-8'))
    return md5.hexdigest()
```

```python
print(type(content))
```

```python
import re
content = str(content)
pattern = '<img .*?src=\"(.*?)\"'
re_img = re.compile(pattern)
```

```python
import os
for imageUrl in re_img.findall(content):
    print(imageUrl)
    imageMd5 = md5(imageUrl)
    filename = 'images/' + imageMd5 + os.path.splitext(imageUrl)[-1]
    print(filename)
    try:
        request.urlretrieve(imageUrl, filename)
    except Exception as e:
        print('图片出错', e)
    else:
        content = content.replace(imageUrl, filename)
```

```python
html_template = """
<!DOCTYPE html>
<html lang="en"><head>
    <meta charset="UTF-8">
</head><body>

<p><center><h1>{title}</h1></center></p>
    
    {content}

</body></html>
"""
```

```python
html = html_template.format(title=title, content=content)
# print(title)
# print(content)
print(html)
```
